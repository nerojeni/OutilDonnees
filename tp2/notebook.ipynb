```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
def visualiser(stock): 
    for k, v in stock.items():
        print(f"Produit : {k} - Quantité : {v} ")
```

```python
visualiser(stock)
```

```python
def fabriquer(prod, quantity):
    stock.update({prod: quantity})
```

```python

fabriquer("chouquettes", 200)
visualiser(stock)
```

```python
tresorie = 0

def update_tresorie(tresorie, price): 
    tresorie += price
    return tresorie

def vendre(prod,quantity) : 
    price = 1 * quantity
    stock[prod] -= quantity 

    update_tresorie(tresorie, price)

    return price
```

```python
vendre("chouquettes", 3)
visualiser(stock)
```
